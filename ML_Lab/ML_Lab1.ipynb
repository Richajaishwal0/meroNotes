```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:1].values
y=df.iloc[:,1:2].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,model.predict(x_train),color="red")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="blue")
plt.plot(x_test,model.predict(x_test),color="red")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import Ridge
# Train Ridge model
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(x_train, y_train)

# Plot
plt.scatter(x_test, y_test, color="blue", label="Actual")
plt.plot(x_test, ridge_model.predict(x_test), color="red", label="Ridge Prediction", linewidth=2)
plt.title("Ridge Regression: Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import Lasso

# Train Lasso model
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(x_train, y_train)

# Plot
plt.scatter(x_test, y_test, color="blue", label="Actual")
plt.plot(x_test, lasso_model.predict(x_test), color="red", label="Lasso Prediction", linewidth=2)
plt.title("Lasso Regression: Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```
